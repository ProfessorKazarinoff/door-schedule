```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
def get_content_div_from_dept_course_page(dept_url="https://www.pcc.edu/schedule/default.cfm?fa=dspTopicDetails&thisTerm=202301&topicid=GE&type=Credit"):
    """
    dept_url = <str>
    returns <bs4.element.Tag>

    """
    page = requests.get(dept_url)
    soup = BeautifulSoup(page.content, "html.parser")
    page_content_div = soup.find(
        "div",
        attrs={"id": "content"},
    )
    return page_content_div
```

```python
page_content_div = get_content_div_from_dept_course_page()
```

```python
type(page_content_div)
```

```python
def get_course_list_dl_from_content_div(page_content_div_bs4_element_Tag):
    dl_bs4_element_Tag = page_content_div_bs4_element_Tag.find('dl', class_="course-list")
    return dl_bs4_element_Tag
```

```python
dl_bs4_element_Tag = get_course_list_dl_from_content_div(page_content_div)
```

```python
type(dl_bs4_element_Tag)
```

```python
def get_list_of_course_links_from_dl(dl_bs4_element_Tag):
    link_lst = []
    dd_bs4_element_Tag = dl_bs4_element_Tag.find_all('dd')
    for dd in dd_bs4_element_Tag:
        a_tag = dd.find("a", href=True)
        link_str = a_tag.get("href")
        link_lst.append(link_str)
    return link_lst
```

```python
get_list_of_course_links_from_dl(dl_bs4_element_Tag)
```

```python
def get_list_of_course_links_from_dept_page(dept_url="https://www.pcc.edu/schedule/default.cfm?fa=dspTopicDetails&thisTerm=202301&topicid=GE&type=Credit"):
    content_div = get_content_div_from_dept_course_page(dept_url)
    dl = get_course_list_dl_from_content_div(content_div)
    link_lst = get_list_of_course_links_from_dl(dl)
    return link_lst
```
